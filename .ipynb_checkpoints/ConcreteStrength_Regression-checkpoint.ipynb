Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
df['age']=df.age.astype(str)
le=LabelEncoder()
df['age']=le.fit_transform(df.age)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X=df.drop('strength',axis=1)
y=df.pop('strength')
```

```python

ss=StandardScaler()
X_scaled=ss.fit_transform(X)
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.30,random_state=1)
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
print("Train Score of simple Linear Model is {}".format(lr_model.score(X_train,y_train)))
print("Test Score of simple Linear Model is {}".format(lr_model.score(X_test,y_test)))
```

```python
from sklearn.metrics import r2_score,mean_squared_error
y_pred=lr_model.predict(X_test)
print("r2_score is {}".format(r2_score(y_test,y_pred)))
print("RMSE is {}".format(np.sqrt(mean_squared_error(y_test,y_pred))))
```
